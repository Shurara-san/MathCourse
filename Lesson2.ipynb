## Задание 1

Так как общее уравнение параболы имеет следующий вид: $y = ax^2 + bx +c$, зная координаты трёх точек, можем составить систему вида:

\begin{equation*}
	\begin{cases}
		a + b + c = 2  \ \ \ \ \ \ \ \ \ \ \ \ \ \ (1) 
		\\
		9a + 3b + c = 10  \ \ \ \ \ \ \ \ (2)
		\\
		25a +5b + c  \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ (3)
	\end{cases}
\end{equation*}

Тогда, из (1) $\Rightarrow c = 2 - a - b \Rightarrow^{(2)} 10 = 9a + 3b + 2 - a - b \Rightarrow b = 4 - 4a$

Подставляем $c$ и $b$ в (3): $1 = 25a + (4 - 4a) + (2 - a - (4 - 4a)) \Rightarrow -1 = 24a \Rightarrow a = -\frac{1}{24}$

Тогда: $b = 4 - \frac{1}{6} = \frac{23}{6}, c = 2 + \frac{1}{24} - 4 + \frac{4}{24} = -\frac{43}{24}$

Для проверки подставим $x = 1: y = -\frac{1}{24} + \frac{23}{6} - \frac{43}{24} = \frac{-1 + 92 - 43}{24} = \frac{48}{24} = 2$

#### Уравнение параболы, проходящей через точки (1,2), (3,10) и (5,1): $y = -\frac{1}{24}x^2 + \frac{23}{6}x - \frac{43}{24}$

## Задание 2

Известно, что свежий огурец на 99% состоит из воды. Месяц назад взвесили мешок со свежими огурцами. Получилось, что мешок весил ровно 100 кг. Мешок убрали, а через месяц снова взвесили. Огурцы за это время усохли, и теперь вода составляет уже только 98% их веса. Сколько теперь (в кг) весят огурцы?

Из первого взвешивания известно, что на 100кг свежих огурцов приходилось 99кг воды (99%). Значит, на всё остальное оставался 1кг, который был равен одному проценту. Так как известно, что после второго взвешивания вода составляла 98% от общей массы огурцов, а вес остальной части измениться не должен был, можно получить, что 2% усохших огурцов равняется 1кг. Следовательно, 0.5кг - вес 1% мешка усохших огурцов. Значит, теперь огурцы весят 50кг.

#### Ответ: 50кг

#### Ответ: $x$ $\approx$ 8,228

### 3. $\log_8{2^{8x-4}} = 4 \Rightarrow \frac{\log_2{2^{8x-4}}}{\log_2{8}} = 4 \Rightarrow \frac{8x-4}{3} = 4 \Rightarrow 8x - 4 = 12 \Rightarrow x = 2$

#### Ответ: $x$ = 2

### 4. $3^{\log_9(5x-5)} = 5$

$3^{\frac{1}{2}\log_3(5x-5)} = 3^{\log_3{\sqrt{5x-5}}}$

$\sqrt(5x-5) = 5 \Rightarrow |5x-5| = 25 \Rightarrow 5x-5 = 25 \Rightarrow x = 6$

(в случае $5x-5 = -25$, получаем отрицательный аргумент в логарифме при положительном основании($\log_9{-25}$), что невозможно)

#### Ответ: $x$ = 6

### 5. $x^{\log_3{x+1}} = 9$

$\log_3{x^{\log_3{x+1}}} = log_3{9} \Rightarrow (log_3{x} + 1)log_3{x} = 2$

Пусть $t = log_3{x}: t^2 + t - 2 = 0 \Rightarrow t_{1,2} = 1, -2$

$t_1 = 1: x = 3, t_2 = -2: x = \frac{1}{9}$

#### Ответ: $x_1$ = 3, $x_2$ = 1/9
